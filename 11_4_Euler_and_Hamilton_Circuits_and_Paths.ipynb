## Some definitions

Note: **"Euler" is pronounced "Oiler".**

An $\underline{E}$uler circuit in a graph $G$ is a simple circuit containing every $\underline{e}$dge of $G$.

An $\underline{E}$uler path in a graph $G$ is a simple path containing every $\underline{e}$dge of $G$.

A Hamil$\underline{t}$on circuit is a circuit that traverses each ver$\underline{t}$ex in $G$ exactly once.

A Hamil$\underline{t}$on path is a path that traverses each ver$\underline{t}$ex in $G$ exactly once.

## Some useful theorems

Note: Recall that *iff* means *if and only if*.

A connected multigraph with at least two vertices has an Euler circuit **iff** each vertex has even degree.

A connected multigraph has an Euler path **iff** it has exactly two vertices of odd degree.

Dirac's Theorem:

If (but **NOT** only if) $G$ is connected, simple, has $n \ge 3$ vertices, and $\forall v$ deg($v$) $\ge \frac{n}{2}$, then $G$ has a Hamilton circuit.

Ore's Theorem:

If (but **NOT** only if) $G$ is connected, simple, has $n \ge 3$ vertices, and $\forall u, v$ (nonadjacent vertices in $G$) deg($u$) $+$ deg($v$) $\ge n$, then $G$ has a Hamilton circuit.
