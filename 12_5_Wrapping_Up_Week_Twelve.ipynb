In the hierarchy of power --- and by power we mean the ability to recognize larger classes of languages --- DFA and NFA are at the bottom, PDA are strictly more powerful than DFA or NFA, and TMs are at the top. Nothing is more powerful than a Turing Machine --- in the context of serial computation and classical (not Quantum) computing. Yet there are problems to be solved (and equivalently, languages to be recognized) that are beyond the power of TMs to solve (recognize).

This is not the complete hierarchy --- there is another intermediate model between PDA and TMs, and there is a difference in power between NONdeterministic PDA and Deterministic PDA (NPDA > DPDA) --- but this is a good rough cut.

What is an example of an operation on two context-free languages that does not result in a context-free language?

### Answer

The intersection of these two context-free languages

$\{\mathtt{a}^{m}\mathtt{b}^{m}\mathtt{c}^{n}\mid m, n \ge 0\} \cap \{\mathtt{a}^{m}\mathtt{b}^{n}\mathtt{c}^{n}\mid m, n \ge 0\}$

is the Context-SENSITIVE language

$\{\mathtt{a}^{n}\mathtt{b}^{n}\mathtt{c}^{n}\mid n \ge 0\}$.

## 7

What is another example?

### Answer

The complement of the context-sensitive language

$L_{ww} = \{ww\mid w \in \{0, 1\}^*\}$

is context-free.

(It may come as a surprise that most programming languages are **not** context-free! For instance, in C, we can declare function **prototypes** that can introduce an arbitrary number of arguments. Later, when the function is defined, the same arguments must appear in the same order. $L_{ww}$ captures the essence of such "define/use" structures.)
